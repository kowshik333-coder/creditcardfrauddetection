```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import StandardScaler # Import StandardScaler

# Load dataset
data = pd.read_csv("creditcard.csv")

# Separate legitimate and fraud transactions
legit = data[data['Class'] == 0]
fraud = data[data['Class'] == 1]

# Create balanced dataset
legit_sample = legit.sample(n=len(fraud), random_state=42)

new_dataset = pd.concat([legit_sample, fraud], axis=0)

# Split features and target
X = new_dataset.drop(columns='Class', axis=1)
Y = new_dataset['Class']

# Train-test split
X_train, X_test, Y_train, Y_test = train_test_split(
    X,
    Y,
    test_size=0.2,
    stratify=Y,
    random_state=2
)

# Scale the data
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

# Train Logistic Regression model
model = LogisticRegression(max_iter=5000)

model.fit(X_train, Y_train)

# Evaluate model
train_pred = model.predict(X_train)
test_pred = model.predict(X_test)

print("Training Accuracy:", accuracy_score(Y_train, train_pred))
print("Testing Accuracy:", accuracy_score(Y_test, test_pred))

# -----------------------------
# USER INPUT PREDICTION
# -----------------------------

print("\nEnter 30 values separated by commas")
print("(Time, V1, V2, ..., V28, Amount)")

user_input = input("Input: ")

try:
    user_data = [float(x.strip()) for x in user_input.split(',')]

    if len(user_data) != 30:
        print("\nError: Please enter exactly 30 values.")
    else:
        # Convert to DataFrame with feature names
        user_df = pd.DataFrame([user_data], columns=X.columns) # X from before scaling

        # Scale user input data using the same scaler
        user_df_scaled = scaler.transform(user_df)

        prediction = model.predict(user_df_scaled)
        probability = model.predict_proba(user_df_scaled)

        print("\nPrediction Result")

        if prediction[0] == 0:
            print("Transaction is LEGITIMATE")
        else:
            print("Transaction is FRAUDULENT")

        print(f"Legitimate Probability: {probability[0][0]:.4f}")
        print(f"Fraud Probability: {probability[0][1]:.4f}")

except ValueError:
    print("\nInvalid input. Please enter only numeric values.")
```
